```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


```

```python
df=pd.read_csv("dataset.csv")
```

```python
df.head()
```

# 1 Histogram

```python
plt.figure(figsize=(5,5))      #set the figure size

#historam
sns.histplot(df["age"],kde=True,bins=30,color="blue",alpha=0.7)
plt.title("Age Distribution")                                     #kde--->kernel denisty estimation
plt.xlabel("age")
plt.ylabel("Frequency")
plt.grid(True,alpha=1)
plt.xlim(0,100)
plt.ylim(0,40)

plt.show()
```

# 2 Scatter plot

```python
# correaltion we use
# to find the relation blw two variables
#to find outliers
#spread of data
#variance of data

plt.figure(figsize=(5,5))                  #set figure size of the graph
sns.scatterplot(x="age",y="chol",data=df,color="blue",edgecolor="black",alpha=0.8,s=100,marker="*")
plt.title("Age Vs Cholesterol")
plt.xlabel("age")
plt.ylabel("cholesterol")
```

```python
plt.figure(figsize=(5,5))                  #set figure size of the graph
sns.scatterplot(x="trestbps",y="chol",data=df,color="blue",edgecolor="black",alpha=0.8,s=100,marker="1")
plt.title("Blood Pressure Vs cholesterol")
plt.xlabel("Blood Pressure")
plt.ylabel("cholesterol")
```

# 3 Line Plot

```python
#trends
#pattern
#series over time

plt.figure(figsize=(5,5))               #set figure size of the graph
sns.lineplot(x="age",y="chol",data=df,color="blue",alpha=0.8,marker="*",markersize=10,linewidth=3)
plt.title("Age Vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.grid(True)
plt.show()
```

# 4 Box Plot

```python
#boxplot
#it is used to find outliers

plt.figure(figsize=(5,5))

sns.boxplot(x="chol",data=df,color="blue",width=0.5)
plt.title("boxplot of cholesterol")
plt.xlabel("cholesterol")
plt.grid(True)
plt.show()
```

# 5 Pie Chart


```python
#pie chart
#proportions

plt.figure(figsize=(5,5))
plt.pie(
    df["target"].value_counts(),
    labels=["No Disease","Disease"],
    colors=["blue","red"],
    autopct="%1.1f%%",
    shadow=True,
    wedgeprops={"edgecolor":"black","linewidth":1},
    textprops={"fontsize":14,"color":"black"}   
)
plt.title("Distribution of target variable")
plt.show()
```

# 5 Bar  Chart

```python
#barchart
#comparision of categories
# to find the frequency of categories
#to find the distribution of categories

plt.figure(figsize=(5,5))
sns.barplot(x="sex",y="age",data=df,edgecolor="black",color="blue",alpha=0.7,width=0.5)
plt.title("Average age by sex")
plt.xlabel("Sex")
plt.ylabel("Average Age")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,6))

sns.set_style("whitegrid")   # cleaner background
sns.set_context("talk")      # larger fonts

ax = sns.barplot(
    x="sex",
    y="age",
    data=df,
    palette="Set2",          # better colors
    edgecolor="black",
    linewidth=1.5,
    width=0.5
)

plt.title("Average Age by Sex", fontsize=16, fontweight='bold')
plt.xlabel("Sex", fontsize=13)
plt.ylabel("Average Age", fontsize=13)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```
