```python
import pandas as pd
```

```python
df=pd.read_csv('Dataset1.csv')
print(df.head())
```

# 1.identify all rows where order_amount is non-numeric but is_delivered is marked as delivered

```python
# #1.identify all rows where order_amount is non-numeric but is_delivered is marked as delivered
# df['order_amount'] = pd.to_numeric(df['order_amount'], errors='coerce').astype('float')
# df['order_amount'].isna()
```

```python

# df['is_delivered']=df['is_delivered'].replace({"yes": "true", "no": "false", True: "true", False: "false"})
# df['is_delivered']
```

```python
df["order_amount"]=pd.to_numeric(df['order_amount'], errors='coerce').isna()
df["is_delivered"]=df["is_delivered"].astype(str).str.lower().isin(["true", "yes"])
df
```

# 2.convert order_date to datetime and find orders where delivery_date is negative after valid date conversion

```python
#2 convert order_date to datetime and find orders where delivery_date is negative after valid date conversion
df['order_date']=pd.to_datetime(df['order_date'], errors='coerce', dayfirst=True)
df['order_date']
```

```python
import numpy as np
df['delivery_days']=pd.to_numeric(df['delivery_days'], errors='coerce')
df.loc[df['delivery_days']<0, 'delivery_days'] = np.nan
df['delivery_days']
```

```python
df
```

```python
df[(df['delivery_days'].notna()) & (df['delivery_days']<0)]
df
```

# 3. standardize payment_mode and calculate the percentage share of each mode per city


```python
#3. standardize payment_mode and calculate the percentage share of each mode per city
# a=df['payment_mode'].value_counts(normalize=True)*100
# a

df['payment_mode']=df['payment_mode'].str.lower().str.strip()
df['city']=df['city'].str.lower().str.strip()
a=df.groupby('city')['payment_mode'].value_counts(normalize=True)*100
a

```

```python
counts = df.groupby(['city', 'payment_mode']).size()

total_per_city = counts.groupby(level=0).sum()

percentage = (counts / total_per_city) * 100

percentage_df= percentage.reset_index(name='percentage')
percentage_df
```

# 4.detect customers whose customer_age is invalid but have order amounts above the dataest median

```python
# 4.detect customers whose customer_age is invalid but have order amounts above the dataest median
df['customer_age']=pd.to_numeric(df['customer_age'], errors='coerce')
df['order_amount']=pd.to_numeric(df['order_amount'], errors='coerce')
m=df['order_amount'].median()
df[(df['customer_age'].isna()) & (df['order_amount']>m)]
```

# 5. find duplicate customer_ids where the same customer placed orders in multiple cities


```python
#5. find duplicate customer_ids where the same customer placed orders in multiple cities
#using groupby
unique_city=df.groupby('customer_id')['city'].nunique()

cust_order=unique_city[unique_city>=1].reset_index()
cust_order
```

#  6.create a flag for orders where delivery_days is missing but is_delivered is true

```python
# # 6.create a flag for orders where delivery_days is missing but is_delivered is true
df['delivery_flag'] = ((df['delivery_days'].isna()) & (df['is_delivered'] == True))
df[['delivery_flag','is_delivered','delivery_days']]
```

# 7.after cleaning,compute the average order value per payment mode,excluding invalid amounts.


```python
#7.after cleaning,compute the average order value per payment mode,excluding invalid amounts.

avgorder=df.groupby('payment_mode')['order_amount'].mean().reset_index(name='average_order_value')
avgorder
```

# 8.identify the cities where more than 30% of orders have missing order_date

```python
#8.identify the cities where more than 30% of orders have missing order_date
total_orders=df.groupby('city').size()  
missing_order=df[df['order_date'].isna()].groupby('city').size()
percentage_mis=(missing_order/total_orders)*100
above30=percentage_mis[percentage_mis>30].reset_index(name='missing_order_date_percentage')
above30
```
