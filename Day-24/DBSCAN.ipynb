```python
import pandas as pd


data = {
       "X": [1,2,2,3,8,8,25,26,27,100],
       "Y": [1,2,3,2,8,9,8,25,27,100]  
}
data=pd.DataFrame(data)
data 
```

```python
#visualize the data
import matplotlib.pyplot as plt
plt.scatter(data["X"],data["Y"])
plt.show()
```

```python
eps1=2
min_points1=3
from sklearn.cluster import DBSCAN
model=DBSCAN(eps=eps1,min_samples=min_points1)
model.fit(data)
```
