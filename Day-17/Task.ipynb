```python
Q1=df['income'].quantile(0.25)
Q3=df['income'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers=df[(df['income']<lower_bound) | (df['income']>upper_bound)]
len(outliers)
```

```python
def cap_outlier(column):
    Q1=df[column].quantile(0.25)
    Q3=df[column].quantile(0.75)
    IQR=Q3-Q1
    Lower = Q1 - 1.5*IQR
    Upper = Q3 + 1.5*IQR
    df[column]=df[column].clip(Lower,Upper)
```

```python
cap_outlier('income')
cap_outlier('loan_amount')
cap_outlier('credit_score')
cap_outlier('num_transactions')
cap_outlier('annual_spend')
```

```python
#handle the categoricla variables by one hot encoding
cat=['city','employment_type','loan_type']
from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder(sparse_output=False)

encoded_data=encoder.fit_transform(df[cat])

encoded_data

```

```python
#handle the scale data
from sklearn.preprocessing import MinMaxScaler
num_cols=['income','loan_amount','credit_score','num_transactions','annual_spend']
scaler=MinMaxScaler()
df[num_cols]=scaler.fit_transform(df[num_cols])
df.head()

```

```python
from sklearn.model_selection import train_test_split
x=df.drop(columns=['target'])
y=df['target']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#model 1

from sklearn.neighbors import KNeighborsRegressor

knn_regressor = KNeighborsRegressor(n_neighbors=5)

knn_regressor.fit(x_train,y_train)

y_pred=knn_regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score

mse=mean_squared_error(y_test,y_pred)

r2=r2_score(y_test,y_pred)

print("mean squared error:",mse)
print("r2 score:",r2)
```

```python
mse_values=[]
k_values=range(1,21)

for k in k_values:
    knn_regressor=KNeighborsRegressor(n_neighbors=k)
    
    knn_regressor.fit(x_train,y_train)
    
    y_pred=knn_regressor.predict(x_test)
    
    mse=mean_squared_error(y_test,y_pred)
    
    mse_values.append(mse)
```

```python
# plot the elbow plot
plt.figure(figsize=(10,5))
plt.plot(k_values, mse_values, marker='o',color='red',linewidth=2,markersize=8,label='MSE',)
plt.title("Elbow Plot for KNN Regression")
plt.xlabel("K values")
plt.ylabel("Mean Squared Error")
plt.legend()
plt.show()
```
